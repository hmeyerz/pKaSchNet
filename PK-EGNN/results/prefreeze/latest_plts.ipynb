```python
import torch
import numpy as np
import glob
p=glob.glob("../../../data/all_checkpoints/*benchmark*.pt")
for c in p[-10:]:
    print(c)
    a=torch.load(c, map_location=torch.device('cpu'))
    c=a["config"]
    
    plosses=a["train_history"]
    vlosses=a["val_history"]
    from matplotlib import pyplot as plt
    #plt.plot(np.array([plosses]).flatten())
    #plt.show()
    plt.plot(np.array([plosses]).flatten(),label="train")

    #plt.title(f"nhr")
    #plt.show()

    l=(len(np.array([plosses]).flatten()))
    from matplotlib import pyplot as plt
    #plt.plot(np.array([plosses]).flatten())
    #plt.show()
    plt.plot(np.array([vlosses]).flatten(), label="val",alpha=0.7)
    plt.xlim(l-100,l+.1)
    plt.ylim(0.2,.9)

    plt.title(("num_ins:", c["dataset"][1],"dim, depth, numrbf:",c["dim"],c["depth"],c["basis"],"drop:",c["dropout"]))
    plt.legend()
    plt.show()
    
```
