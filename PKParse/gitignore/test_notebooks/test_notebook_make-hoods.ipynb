```python

from sklearn.neighbors import NearestNeighbors
import numpy as np #after debug
dat=np.load("../inputs/1a8d.npz",allow_pickle=True)
z       = dat["z"]           # (N,)
pos     = dat["pos"]         # (N,3)
#seglen  = dat["seglen"]      # (S,)
anchor  = dat["sites"]      # (S,)
ids     = dat["ids"]         # (S,)
pks     = dat["pks"]         # (S,)
n_neighbors=200
nbrs = NearestNeighbors(n_neighbors=n_neighbors, algorithm="brute")
def hoods(all_feats, all_coors,site_coors,n_neighbors):
        "as np arrays"
        neigh = nbrs.fit(all_coors,n_neighbors)
        neighbors = neigh.kneighbors(site_coors,n_neighbors, return_distance=False)
        out=[all_feats[n_ixs] for n_ixs in neighbors], [all_coors[n_ixs].astype(np.float32) for n_ixs in neighbors]
        return out#[all_feats[n_ixs] for n_ixs in neighbors], [all_coors[n_ixs].astype(np.float32) for n_ixs in neighbors]
out=hoods(z,pos,anchor,n_neighbors)  
```
